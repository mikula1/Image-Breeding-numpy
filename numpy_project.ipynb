```python
import numpy as np
import cv2 as cv
from google.colab.patches import cv2_imshow
```

```python
#Loading the image
img = cv.imread('dog.jpg')
print(type(img))
print(img.shape, "\n----------------------")

# Uncomment the line below for non-grayscale image conversion
# img = cv.imread('dog.jpg')

img = cv.imread('dog.jpg', 0) #And comment this for RGB image
print(img.shape)
```

```python
cv2_imshow(img)
```

```python
#Cropping the image to even pixels
sub_img = img[:img.shape[0] - img.shape[0] % 2, :img.shape[0] - img.shape[0] % 2]
sub_img.shape
```

```python
#Splitting the image to 1 pixel width strips from left to right and assigning them to separate images
img_split_1, img_split_2 = sub_img[:,0::2], sub_img[:,1::2]

#Joining the two images by 1 axes (by heigth)
img_split = np.concatenate((img_split_1, img_split_2),axis=1)
cv2_imshow(img_split)
```

```python
#Splitting the image to 1 pixel heigth strips from top to bottom
img_split_1, img_split_2 = img_split[0::2,:], img_split[1::2,:]
print(img_split_1.shape)

#Joining the two images by 0 axes (by width)
cv2_imshow(np.concatenate((img_split_1, img_split_2)))
```
